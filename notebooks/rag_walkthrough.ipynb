# GenAI Production Starter Kit — RAG Walkthrough

This notebook walks through the core components of the RAG pipeline,
agent routing, and security guardrails using the `src/` modules.

## 1. Setup

```python
import sys
sys.path.insert(0, '..')

from src.rag.retriever import retrieve_documents, generate_response
from src.security.input_validation import is_prompt_safe
from src.security.pii_redaction import redact_pii
from src.agents.agent_router import route_query
```

## 2. Input Validation

```python
query = "What is the remote work policy?"

if is_prompt_safe(query):
    print("Query is safe. Proceeding...")
else:
    print("Unsafe query detected. Blocking request.")
```

## 3. PII Redaction

```python
raw_text = "Contact John at john@example.com or call 555-867-5309"
redacted = redact_pii(raw_text)
print("Original:", raw_text)
print("Redacted:", redacted)
```

## 4. RAG Pipeline

```python
docs = retrieve_documents(query)
print("Retrieved documents:")
for doc in docs:
    print(f"  - {doc}")

answer = generate_response(query, docs)
print("\n" + answer)
```

## 5. Agent Routing

```python
test_queries = [
    "What is the remote work policy?",
    "Summarize the Q3 report.",
    "Extract action items from the meeting notes.",
]

for q in test_queries:
    result = route_query(q)
    print(f"Query: {q}")
    print(f"  Intent: {result['intent']} | Response: {result['response']}\n")
```
